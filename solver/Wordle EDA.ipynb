```python
import helpers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
solutions = helpers.get_solution_words()
all_words = helpers.get_all_words()
solutions
```

Histogram of Letters in Solution Words

```python
hist = dict()
for word in solutions:
    seen_in_word = set()
    for letter in word:
        if letter not in seen_in_word:
            seen_in_word.add(letter)
            if letter not in hist:
                hist[letter] = 1
            else:
                hist[letter] = hist[letter] + 1
        
```

```python
hist
```

```python
df = pd.DataFrame(hist.items(), columns = ['letter','count']) 
sorted_df = df.sort_values(by='count')
plt.bar(sorted_df['letter'], sorted_df['count'], width=.8, color='g')
```

```python
best_letters = ['e','a','r', 'o', 't']

def init_positional_hist():
    position_hist = dict()
    for letter in best_letters:
        position_hist[letter] = dict()
        for i in range(1,6):
            position_hist[letter][i] = 0
    return position_hist

def calculate_positional_hist():
    position_hist = init_positional_hist()
    for word in solutions:
        for i, letter in enumerate(word):
            if letter in best_letters:
                position_hist[letter][i+1] += 1
    return position_hist

position_hist = calculate_positional_hist()
print(position_hist)
# set width of bar
barWidth = 0.15
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
e = position_hist['e'].values()
a = position_hist['a'].values()
r = position_hist['r'].values()
o = position_hist['o'].values()
t = position_hist['t'].values()
 
# Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
br5 = [x + barWidth for x in br4]

    
# Make the plot
plt.bar(br1, e, color ='r', width = barWidth,
        edgecolor ='grey', label ='e')
plt.bar(br2, a, color ='orange', width = barWidth,
        edgecolor ='grey', label ='a')
plt.bar(br3, r, color ='yellow', width = barWidth,
        edgecolor ='grey', label ='r')
plt.bar(br4, o, color ='g', width = barWidth,
        edgecolor ='grey', label ='o')
plt.bar(br5, t, color ='blue', width = barWidth,
        edgecolor ='grey', label ='t')
 
# Adding Xticks
plt.xlabel('Position In Word', fontweight ='bold', fontsize = 15)
plt.ylabel('Number of Occurences', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))],
        ['1', '2', '3', '4', '5'])
 
plt.legend()
plt.show()
```
